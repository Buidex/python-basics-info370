### 2 - Which model is best?

2-1: Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
y = wdbc.diagnosis
X = np.hstack((wdbc['concpoints.mean'].values.reshape((-1, 1)), 
               wdbc['fracdim.mean'].values.reshape((-1, 1))
              ))

# 20% of the data will be used for validation, and 80% for training
Xtrain, Xvalid, ytrain, yvalid = train_test_split(X, y, test_size=0.2)
```

2-2: Fitting the model on the _training_ data:

```python
m = LogisticRegression(solver='lbfgs').fit(Xtrain, ytrain)
```

2-3: Predicting the outcome on training data and creating a confusion matrix of the result

```python
y_hat_train = m.predict(Xtrain)
conf_matrix = pd.crosstab(ytrain, y_hat_train)
conf_matrix
```

2-4: Calculating **accuracy** and the **F-score**

```python
# Accuracy (percentage of correct answers = (TP + TN) / N )
acc = (conf_matrix[0][0] + conf_matrix[1][1]) / (conf_matrix[0][0] + conf_matrix[0][1] + conf_matrix[1][0] + conf_matrix[1][1])
```

```python
# Precision (TP / (TP + FP)) = (TP / predicted positive)
prec = conf_matrix[0][0] / (conf_matrix[0][0] + conf_matrix[0][1])
```

```python
# Recall (percentage of real positives identified) = (TP / (TP + FN))
rec = conf_matrix[0][0] / (conf_matrix[0][0] + conf_matrix[1][0])
```

```python
# F score (harmonic mean of Precision and Recall) = 2 / (1/precision + 1/recall)
f_score = 2 / (1/prec + 1/rec)
```

```python
print(f'Accuracy: {round(acc, 2)} ; F-score: {round(f_score, 2)}')
```

2-5: Predicting the outcome on validation data and creating a confusion matrix of the result

```python
y_hat_valid = m.predict(Xvalid)
conf_matrix_valid = pd.crosstab(yvalid, y_hat_valid)
conf_matrix_valid
```

2-6: Calculating **accuracy** and the **F-score** for the validation data

```python
# Accuracy (percentage of correct answers = (TP + TN) / N )
acc_valid = (conf_matrix_valid[0][0] + conf_matrix_valid[1][1]) / \
        (conf_matrix_valid[0][0] + conf_matrix_valid[0][1] + conf_matrix_valid[1][0] + conf_matrix_valid[1][1])
```

```python
# Precision (TP / (TP + FP)) = (TP / predicted positive)
prec_valid = conf_matrix_valid[0][0] / (conf_matrix_valid[0][0] + conf_matrix_valid[0][1])
```

```python
# Recall (percentage of real positives identified) = (TP / (TP + FN))
rec_valid = conf_matrix_valid[0][0] / (conf_matrix_valid[0][0] + conf_matrix_valid[1][0])
```

```python
# F score (harmonic mean of Precision and Recall) = 2 / (1/precision + 1/recall)
f_score_valid = 2 / (1/prec_valid + 1/rec_valid)
```

```python
print(f'Accuracy: {round(acc_valid, 2)} ; F-score: {round(f_score_valid, 2)}')
```

2-7: We had higher accuracy and F-score for the testing data than for the training data, which **should not happen**. Even though these differences are not so big - about 4% for each of these metrics, it is still important to take note of them.

Because we used different data to be trained compared to the one used for validation and still had these inconsistencies, we have an **indicator of overfitting**. This might be due to the small sample we have available.
